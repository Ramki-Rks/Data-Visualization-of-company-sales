```python
def longest_substring(strings,k):
  output={}
  index=[]
  for i in strings:
    if i in output:
      output[i]+=1
    else:
      output[i]=+1
      #if getitng distinct k character then excute this block
      if (output.get(i))==k:
        for j in strings:
          if i==j:
            index.append(strings.find(j))
          else:
            continue
      else:
        continue

    
    return index
```
